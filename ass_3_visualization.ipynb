```
path="https://raw.githubusercontent.com/prashant90066/18CSE049/main/Bank_churn_modelling.csv"
```

```
import pandas as pd
```

```
import numpy as np
```

```
import matplotlib.pyplot as plt
```

```
data = pd.read_csv(path)
```

```
data
```

```
data.dropna(axis=0,inplace=True)
```

```
data.shape
```

```
data.head()
```

```
plt.scatter(data['Age'],data['Balance'],c='green')
plt.xlabel('Age(years)')
plt.ylabel('Balance	')
plt.show()
```

```
plt.hist(data['Balance'])
plt.show()
```

```
plt.hist(data['Balance'],color='green',edgecolor='white',bins=5)
plt.title("histogram of balance")
plt.xlabel("frequency")
plt.ylabel("balance")
plt.show()
```

```
import seaborn as sns
```

```
sns.set(style='darkgrid')
sns.regplot(x=data['Age'],y=data['Balance'])
```

```
sns.set(style='darkgrid')
sns.regplot(x=data['Age'],y=data['Balance'],marker='*',fit_reg=False)
```

```
sns.distplot(data['Age'])
```

```
sns.distplot(data['Age'],kde=False)
```

```
sns.distplot(data['Age'],kde=False,bins=5)
```

```
sns.countplot(x="Geography",data=data)
```

```
sns.countplot(x="Geography",data=data,hue='IsActiveMember')
```

```
sns.boxplot(y=data['Balance'])
```
